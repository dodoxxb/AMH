```python
import os
import cv2 as cv
import pandas as pd
import numpy as np
```

```python
def get_face_vari(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_face = df['face_vari'].values.tolist()
    
    return df_face

def get_face(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_face = df['face'].values.tolist()
    
    return df_face

def get_iris(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_iris = df['iris'].values.tolist()
    
    return df_iris

def get_physical_index(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_index = df['physical_index'].values.tolist()

    return df_index

def get_voice_feature(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_mfcc = df['mfcc'].values.tolist()
    
    return df_mfcc

def get_wav2vec(csv_path):
    df = pd.read_csv(csv_path, usecols=[1])
    df_wav2vec = df['wav2vec'].values.tolist()
    
    return df_wav2vec

def get_context_feature(csv_path):
    df = pd.read_csv(csv_path, usecols=[1,2])
    df_pert = df['pert'].values.tolist()
    df_roberta = df['roberta'].values.tolist()
    return df_pert, df_roberta

if __name__ == '__main__':
    face = []
    face_vari = []
    iris = []
    physical_index = []
    mfcc_ft = []
    wav2vec = []
    pert = []
    roberta = []
    macberta = []
    # con_pert = []
    # con_roberta = []
    # con_macberta = []
    graph = []
    sample_index = 0

    root = os.path.dirname(os.getcwd())
    print("root direction", root)
    dataset = os.listdir(root + '/muldataset')
    del dataset[-2:]
    for sample in dataset:
        # if '00002' in sample:
        #     continue
        sample_path = root + '/muldataset' + '/' + sample #
        # print(sample_path)
        sample_file = os.listdir(sample_path) #
        # print(sample_file)
        for detail in sample_file:
                sample_path = root + '/muldataset' + '/' + sample
                if 'fft_face.csv' in detail:
                    # csv
                    print(sample_path + '/' + detail)
                    temp_face = get_face(sample_path + '/' + detail)
                    face.append(temp_face)
                    print(np.array(face).shape)

                if 'fft_face_vari.csv' in detail:
                    # csv
                    print(sample_path + '/' + detail)
                    temp_face = get_face_vari(sample_path + '/' + detail)
                    face_vari.append(temp_face)
                    print(np.array(face_vari).shape)

                if 'fft_iris' in detail:
                    print(sample_path + '/' + detail)
                    temp_iris = get_iris(sample_path + '/' + detail)
                    iris.append(temp_iris)
                    print(np.array(iris).shape)

                if 'fft_physical_index' in detail:
                    print(sample_path + '/' + detail)
                    temp_index = get_physical_index(sample_path + '/' + detail)
                    physical_index.append(temp_index)
                    print(np.array(physical_index).shape)

                if 'fft_mfcc' in detail:
                    print(sample_path + '/' + detail)
                    temp_mfcc = get_voice_feature(sample_path + '/' + detail)
                    mfcc_ft.append(temp_mfcc)
                    print(np.array(mfcc_ft).shape)

                if 'fft_wav2vec' in detail:
                    print(sample_path + '/' + detail)
                    temp_wav2vec = get_wav2vec(sample_path + '/' + detail)
                    wav2vec.append(temp_wav2vec)
                    print(np.array(wav2vec).shape)

                if 'fft_text' in detail:
                    print(sample_path + '/' + detail)
                    temp_pert, temp_roberta = get_context_feature(sample_path + '/' + detail)
                    pert.append(temp_pert)
                    roberta.append(temp_roberta)
                    print(np.array(pert).shape, np.array(roberta).shape)

        sample_index += 1        

    print("Face set:{face}".format(face = np.array(face).shape))
    print("Iris set:{iris}".format(iris = np.array(iris).shape))
    print("Physical index set:{index}".format(index = np.array(physical_index).shape))
    print("MFCC set:{mfcc}".format(mfcc = np.array(mfcc_ft).shape))
    print("Wav2vec set:{wav}".format(wav = np.array(wav2vec).shape))
    print("Graph set:{gra}".format(gra = np.array(graph).shape))
    print("Pert set:{pert}, Roberta:{robert}".format(pert = np.array(pert).shape, \
        robert = np.array(roberta).shape))
```

```python
from dtaidistance import dtw
from sklearn import preprocessing
def graph_feature(face, face_vari, iris, physical_index, mfcc, wav2vec, pert, roberta):
    graph = []

    # face = face[0]
    # iris = iris[0]
    # physical_index = physical_index[0]
    # mfcc = mfcc[0]
    # pert = pert[0]
    # roberta = roberta[0]
    # macbert = macbert[0]

    face = np.array(face)
    face_vari = np.array(face_vari)
    iris = np.array(iris)
    physical_index = np.array(physical_index)
    mfcc = np.array(mfcc)
    wav2vec = np.array(wav2vec)
    pert = np.array(pert)
    roberta = np.array(roberta)
    
    graph.append(dtw.distance_fast(face, face))
    graph.append(dtw.distance_fast(face, face_vari))
    graph.append(dtw.distance_fast(face, iris))
    graph.append(dtw.distance_fast(face, physical_index))
    graph.append(dtw.distance_fast(face, mfcc))
    graph.append(dtw.distance_fast(face, wav2vec))
    graph.append(dtw.distance_fast(face, pert))
    graph.append(dtw.distance_fast(face, roberta))
    
    graph.append(dtw.distance_fast(face_vari, face))
    graph.append(dtw.distance_fast(face_vari, face_vari))
    graph.append(dtw.distance_fast(face_vari, iris))
    graph.append(dtw.distance_fast(face_vari, physical_index))
    graph.append(dtw.distance_fast(face_vari, mfcc))
    graph.append(dtw.distance_fast(face_vari, wav2vec))
    graph.append(dtw.distance_fast(face_vari, pert))
    graph.append(dtw.distance_fast(face_vari, roberta))

    graph.append(dtw.distance_fast(iris, face))
    graph.append(dtw.distance_fast(iris, face_vari))
    graph.append(dtw.distance_fast(iris, iris))
    graph.append(dtw.distance_fast(iris, physical_index))
    graph.append(dtw.distance_fast(iris, mfcc))
    graph.append(dtw.distance_fast(iris, wav2vec))
    graph.append(dtw.distance_fast(iris, pert))
    graph.append(dtw.distance_fast(iris, roberta))

    graph.append(dtw.distance_fast(physical_index, face))
    graph.append(dtw.distance_fast(physical_index, face_vari))
    graph.append(dtw.distance_fast(physical_index, iris))
    graph.append(dtw.distance_fast(physical_index, physical_index))
    graph.append(dtw.distance_fast(physical_index, mfcc))
    graph.append(dtw.distance_fast(physical_index, wav2vec))
    graph.append(dtw.distance_fast(physical_index, pert))
    graph.append(dtw.distance_fast(physical_index, roberta))

    graph.append(dtw.distance_fast(mfcc, face))
    graph.append(dtw.distance_fast(mfcc, face_vari))
    graph.append(dtw.distance_fast(mfcc, iris))
    graph.append(dtw.distance_fast(mfcc, physical_index))
    graph.append(dtw.distance_fast(mfcc, mfcc))
    graph.append(dtw.distance_fast(mfcc, wav2vec))
    graph.append(dtw.distance_fast(mfcc, pert))
    graph.append(dtw.distance_fast(mfcc, roberta))   

    graph.append(dtw.distance_fast(wav2vec, face))
    graph.append(dtw.distance_fast(wav2vec, face_vari))
    graph.append(dtw.distance_fast(wav2vec, iris))
    graph.append(dtw.distance_fast(wav2vec, physical_index))
    graph.append(dtw.distance_fast(wav2vec, mfcc))
    graph.append(dtw.distance_fast(wav2vec, wav2vec))
    graph.append(dtw.distance_fast(wav2vec, pert))
    graph.append(dtw.distance_fast(wav2vec, roberta))   

    graph.append(dtw.distance_fast(pert, face))
    graph.append(dtw.distance_fast(pert, face_vari))
    graph.append(dtw.distance_fast(pert, iris))
    graph.append(dtw.distance_fast(pert, physical_index))
    graph.append(dtw.distance_fast(pert, mfcc))
    graph.append(dtw.distance_fast(pert, wav2vec))
    graph.append(dtw.distance_fast(pert, pert))
    graph.append(dtw.distance_fast(pert, roberta))

    graph.append(dtw.distance_fast(roberta, face))
    graph.append(dtw.distance_fast(roberta, face_vari))
    graph.append(dtw.distance_fast(roberta, iris))
    graph.append(dtw.distance_fast(roberta, physical_index))
    graph.append(dtw.distance_fast(roberta, mfcc))
    graph.append(dtw.distance_fast(roberta, wav2vec))
    graph.append(dtw.distance_fast(roberta, pert))
    graph.append(dtw.distance_fast(roberta, roberta))

    return graph
```
